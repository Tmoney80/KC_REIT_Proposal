```python
kc_trans_df = pd.get_dummies(kc_df, columns=['waterfront', 'greenbelt', 'nuisance', 'view', 'condition', 'grade',
       'heat_source', 'sewer_system'])
kc_trans_df
```

```python
kc_trans_df.corr()['price'].sort_values()
```

```python
kc_trans_df = pd.get_dummies(kc_df, columns=['waterfront', 'greenbelt', 'nuisance', 'view', 'condition', 'grade',
       'heat_source', 'sewer_system'], drop_first=True)
kc_trans_df
```

```python
X = kc_trans_df['sqft_living']
y = kc_trans_df['price']
base_model = sm.OLS(endog=y, exog=sm.add_constant(X))
base_results = base_model.fit()
base_results.summary()
```

```python
base_results.fvalue, base_results.f_pvalue, base_results.rsquared_adj
```

```python
cols = kc_trans_df.columns
for col in cols:
    X = kc_trans_df[col]
    y = kc_trans_df['price']
    base_model = sm.OLS(endog=y, exog=sm.add_constant(X))
    base_results = base_model.fit()
    print(f'Var = {col}, Fval = {base_results.fvalue}, F_pval = {base_results.f_pvalue}, R2d = {base_results.rsquared_adj}')
    #print("Variable = {}, Fvalue = {}, F_pvalue = {}, R2d = {}".format(col, base_results.fvalue, base_results.f_pvalue, base_results.rsquared_adj))
```

```python

f_vals, fp_vals, adjr_vals = [], [], []
cols = kc_trans_df.columns
for col in cols:
    X = kc_trans_df[col]
    y = kc_trans_df['price']
    base_model = sm.OLS(endog=y, exog=sm.add_constant(X))
    base_results = base_model.fit()
    f_vals.append(base_results.fvalue)
    fp_vals.append(base_results.f_pvalue)
    adjr_vals.append(base_results.rsquared_adj)
    #print(f'Var = {col}, Fval = {base_results.fvalue}, F_pval = {base_results.f_pvalue}, R2d = {base_results.rsquared_adj}')
    #print("Variable = {}, Fvalue = {}, F_pvalue = {}, R2d = {}".format(col, base_results.fvalue, base_results.f_pvalue, base_results.rsquared_adj))
df = pd.DataFrame(list(zip(cols, f_vals)),
               columns =['Var', 'Fval'])
df
```

```python
X = kc_trans_df[['sqft_living', 'bathrooms']]
y = kc_trans_df['price']
mr1_model = sm.OLS(endog=y, exog=sm.add_constant(X))
mr1_results = mr1_model.fit()
mr1_results.summary()
```

```python
sorted_coor_cols = kc_trans_df.corr()['price'].sort_values()

i = len(sorted_coor_cols) - 3
while i > 0:
    X = kc_trans_df[['sqft_living', sorted_coor_cols.index[i]]]
    y = kc_trans_df['price']
    mrl_model = sm.OLS(endog=y, exog=sm.add_constant(X))
    mrl_results = mrl_model.fit()
    print(f'Var = sqft_living & {sorted_coor_cols.index[i]}, Fval = {mrl_results.fvalue}, F_pval = {mrl_results.f_pvalue}, R2d = {mrl_results.rsquared_adj}')
    i -= 1
    
```
